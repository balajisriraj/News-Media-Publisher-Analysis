### Config

```python
#config will allow us to access the specified url for which we are #not authorized. Sometimes we may get 403 client error while parsing #the link to download the article.
nltk.download('punkt')
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent
config.request_timeout = 15

############################################
region = 'IN'
start = datetime.date(2022, 1, 24)
end = datetime.date(2022, 2, 24)
search = 'Adani'
googlenews = gnews.GNews(language='en', start_date=start, end_date=end)
```

```python
list=[]
failed_list = []
for ind in df.index:
    dict={}
    article = googlenews.get_full_article(df[df.index==ind]['url'])
    try:
        article.parse()
        article.nlp()
        dict['Published_Date_raw'] = list((df[df.index==ind]['published date']))[0]
        dict['Published_Date']=str(article.publish_date)
        dict['Title']=article.title
        dict['Article']=article.text
        dict['Summary']=article.summary
        dict['Authors'] = article.authors
        dict['Keywords'] = article.keywords
        dict['href'] = dict(list(df[df.index==ind]['publisher'])[0])['href']
        dict['href_title'] = dict(list(df[df.index==ind]['publisher'])[0])['title']
        list.append(dict)
    except:
        failed_list.append(ind)
    break
# news_df=pd.DataFrame(list)
# print(news_df.shape)
```
